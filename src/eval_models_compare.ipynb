```python
import torch
import torch.nn as nn
import numpy as np
import string

# Фиксируем seed для воспроизводимости
torch.manual_seed(42)
np.random.seed(42)

# Пример текста (можно заменить на любой другой)
import zipfile
from bs4 import BeautifulSoup

def extract_text_from_epub(epub_path):
    """Извлекает текст из EPUB файла.

    Args:
        epub_path: Путь к EPUB файлу.

    Returns:
        Строка, содержащая весь текст из EPUB файла, или None в случае ошибки.
    """
    try:
        with zipfile.ZipFile(epub_path, 'r') as epub_file:
            text = ""
            for filename in epub_file.namelist():
                if filename.endswith(".xhtml") or filename.endswith(".html"):  # Проверяем расширение
                    with epub_file.open(filename) as html_file:
                        html_content = html_file.read().decode('utf-8', errors='ignore') # Декодируем в UTF-8, игнорируем ошибки
                        soup = BeautifulSoup(html_content, 'html.parser')
                        # Извлекаем текст, исключая скрипты и стили
                        for script in soup(["script", "style"]):
                            script.extract()
                        text += soup.get_text(" ", strip=True) + "\n" # Добавляем текст с пробелами между параграфами
            return text
    except FileNotFoundError:
        print(f"Файл {epub_path} не найден.")
        return None
    except zipfile.BadZipFile:
        print(f"Файл {epub_path} поврежден.")
        return None
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return None

# Пример использования
epub_file_path = "ruslan-i-lyudmila.epub"  # Замените на путь к вашему файлу
extracted_text = extract_text_from_epub(epub_file_path)
text = extracted_text

# Словарь: уникальные символы
chars = sorted(set(text))
char_to_idx = {ch:i for i, ch in enumerate(chars)}
idx_to_char = {i:ch for i, ch in enumerate(chars)}

vocab_size = len(chars)
embedding_dim = 8
hidden_size = 32
seq_length = 5  # длина входной последовательности
batch_size = 1
num_layers = 1

# Переводим строку в тензор индексов
def char_to_tensor(text):
    return torch.tensor([char_to_idx[c] for c in text], dtype=torch.long)

# Генерация обучающих пар (input -> target)
def get_batches(text, seq_len):
    data = char_to_tensor(text)
    num_samples = len(data) - seq_len
    for i in range(0, num_samples):
        x = data[i:i+seq_len]
        y = data[i+1:i+seq_len+1]
        yield x.unsqueeze(0), y.unsqueeze(0)

# Определение модели
class CharLSTM(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, num_layers):
        super(CharLSTM, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_size, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_size, vocab_size)

    def forward(self, x, hidden=None):
        x = self.embedding(x)
        out, hidden = self.lstm(x, hidden)
        logits = self.linear(out)
        return logits, hidden

# Инициализируем модель, loss и optimizer
model = CharLSTM(vocab_size, embedding_dim, hidden_size, num_layers)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Обучение
n_epochs = 100
for epoch in range(n_epochs):
    total_loss = 0
    hidden = None
    for x, y in get_batches(text, seq_length):
        logits, hidden = model(x, hidden)
        loss = criterion(logits.view(-1, vocab_size), y.view(-1))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

# Функция генерации текста
def generate_text(model, start_str="h", length=100):
    model.eval()
    generated = start_str
    x = char_to_tensor(start_str).unsqueeze(0)

    with torch.no_grad():
        hidden = None
        for _ in range(length):
            logits, hidden = model(x, hidden)
            last_logits = logits[0, -1, :]
            probs = torch.softmax(last_logits, dim=-1)
            pred_idx = torch.multinomial(probs, 1).item()
            pred_char = idx_to_char[pred_idx]

            generated += pred_char
            x = torch.tensor([[pred_idx]], dtype=torch.long)

    return generated

# Генерация
print("\nСгенерированный текст:")
print(generate_text(model, start_str="h", length=100))
```
